```python
df = pd.DataFrame(
    {
     'Latitude': [34.1588519, 34.0983126, 34.1983126],
     'Longitude': [-118.1668327, -118.2606155, -118.1668327],
     'place-name':['Adoption Center','Favorite Hiking Spot','Favorite Zoomies Spot'],
      'memory':['Where we first met and fell instantly in love <3',
                'Tuna Canyon - Our favorite local park to hike and meet doggie best friends',
                'Zuma - Our favorite place to run free with all the zoomies']})
```

```python
gdf = gpd.GeoDataFrame(
    df, geometry=gpd.points_from_xy(df.Longitude, df.Latitude))
```

```python
gdf.head()
```

Below is an arbitrary field I am creating to help control the size of the icons in the map itself. You can effectively ignore. 

```python
gdf['sizes'] = 2
gdf.sizes
```

You can select different icons from here:

I have chosen hearts for obvious reasons.

```python
m = folium.Map(location=[df.Latitude.mean(), df.Longitude.mean()], zoom_start=12, tiles='cartodbpositron',prefer_canvas=True)
#tile options: cartodbpositron, cartodbdark_matter
# zoom: how zoomed in on your area you want
for index, location_info in df.iterrows():
    icon=folium.Icon(color='black', icon='heart', icon_color="red", prefix='fa')
    # do not touch the prefix unless you can an icon from a difference source
    # color is the background color, icon_color is the color of the icon
    # you can select icons from here: 

    folium.Circle([location_info["Latitude"], location_info["Longitude"]],radius = 100,color='red',fill_color='red').add_to(m)
    folium.Marker([location_info["Latitude"], location_info["Longitude"]],icon =icon,tooltip=location_info['memory']).add_to(m)
folium.TileLayer(
        tiles = 'https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
        attr = 'Esri',
        name = 'Satellite',
        overlay = True,
        control = True
       ).add_to(m)
	
# if you want to add another basemap you can uncomment this 
#folium.TileLayer('cartodbdark_matter').add_to(m)
folium.LayerControl().add_to(m)


m
```
